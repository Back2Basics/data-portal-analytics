```python
# bring in file downloaded from socrata on 2 oct 2016

with open('site_metrics.csv', mode='r') as infile:
    reader = csv.DictReader(infile)
    data = []
    for row in reader:
        data.append(row)
```

```python
# load into dataframe so it's easier to analyze

df = pd.DataFrame(data)
```
